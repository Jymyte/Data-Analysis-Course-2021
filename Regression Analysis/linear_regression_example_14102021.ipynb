```python
import numpy as np
import matplotlib.pyplot as plt
#pip install scikit-learn
from sklearn.linear_model import LinearRegression
```

```python
#Let's create test data
#vectors x and y, both have 100 elements
#Goal is to find regression between x and y

x = np.arange(0,100,1)
y = 3*x + 2 + np.random.randn(100)*30

plt.scatter(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
#x.shape
```

```python
x = x.reshape(-1, 1)
y = y.reshape(-1, 1)
reg = LinearRegression().fit(x,y)
print(reg.score(x,y))

y_reg = reg.predict(x)

plt.figure(figsize = (10,5), dpi = 80)
plt.scatter(x,y)
plt.plot(x,y_reg)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend(['Regression','Original test data'])
plt.show()

#Let's define the regression line
```
