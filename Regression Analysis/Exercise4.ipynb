```python
import pandas as pd
import numpy as np
import plotly.express as px
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression


data = px.data.gapminder().query("continent == 'Europe'")
suomiDF = data[data.country == 'Finland']
```

```python
#Life expectation vs gpd

plt.scatter(suomiDF.gdpPercap, suomiDF.lifeExp)
plt.xlabel('LifeExp')
plt.ylabel('gdp')
plt.show()
```

```python
#Life expectation vs time

plt.scatter(suomiDF.year, suomiDF.lifeExp)
plt.xlabel('Year')
plt.ylabel('gdp')
plt.show()
```

```python
#Life expectation vs GDP fitting score: 0.9635515803066774



gdpReshaped = suomiDF.gdpPercap.values.reshape(-1, 1)
lifeExpReshaped = suomiDF.lifeExp.values.reshape(-1, 1)

reg = LinearRegression().fit(gdpReshaped, lifeExpReshaped)
print(reg.score(gdpReshaped, lifeExpReshaped))

lifeExpReshaped_reg = reg.predict(gdpReshaped)

plt.scatter(gdpReshaped, lifeExpReshaped)
plt.plot(gdpReshaped,lifeExpReshaped_reg)
plt.xlabel('gdp')
plt.ylabel('lifeExp')
plt.grid(True)
plt.legend(['Regression','Original test data'])
plt.show()
```

```python
#Life expectation vs Time fitting score: 0.9938383497403837
#
#As the time has passed so has the life expectancy of the citizens of Finland.
#This can be explained by the increase in gdp to an accuracy of roughly 96%

yearReshaped = suomiDF.year.values.reshape(-1,1)

reg = LinearRegression().fit(yearReshaped, lifeExpReshaped)
print(reg.score(yearReshaped, lifeExpReshaped))

lifeExpReshaped_reg = reg.predict(yearReshaped)

plt.scatter(yearReshaped, lifeExpReshaped)
plt.plot(yearReshaped, lifeExpReshaped_reg)
plt.xlabel('gdp')
plt.ylabel('year')
plt.grid(True)
plt.legend(['Regression','Original test data'])
plt.show()
```
