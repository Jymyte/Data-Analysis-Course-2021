```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array(range(0,100))/10
y = np.sin(x)

fig = plt.figure(num=None, figsize=(12, 4), dpi=100, facecolor='w', edgecolor='k')
ax = fig.add_subplot(1,1,1)
plt.plot(x,y)
plt.xlabel('Time')
plt.ylabel('Signal')
plt.legend(['Sinusoidal'])
plt.grid(True)
ticksx = np.arange(0,10,1)
ticksy = np.arange(-1.1,1.1,0.2)
ax.set_xticks(ticksx)
ax.set_yticks(ticksy)
plt.show()
```

```python
import pandas as pd
data = pd.read_csv('london_crimes.csv')

crime_count = data.groupby(['year'])['number_of_cases'].sum()

fig = plt.figure(num=None, figsize=(10, 3), dpi=100, facecolor='w', edgecolor='k')
plt.bar(crime_count.index,crime_count.values)
plt.show()

fig = plt.figure(num=None, figsize=(10, 3), dpi=100, facecolor='w', edgecolor='k')
plt.plot(crime_count.index,crime_count.values)
plt.ylim([0,800000])
plt.show()
```
