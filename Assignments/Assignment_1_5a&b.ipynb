```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("C:\\Users\\Jyri\\Documents\\Koulu 2021\\SY21 Data Analysis\\Python\\Assignments\\TieliikenneAvoinData_5_14\\TieliikenneAvoinData_5_14.csv", encoding = 'unicode_escape', engine ='python', error_bad_lines=False, delimiter=';', usecols=["merkkiSelvakielinen", "ajoneuvoluokka", "iskutilavuus", "sylintereidenLkm"])
```

```python
print(data.shape[0])
print(len(data))
```

```python
print(data.merkkiSelvakielinen.mode())
data[(data.ajoneuvoluokka == 'M1') & (data.merkkiSelvakielinen == 'Toyota')].groupby(['ajoneuvoluokka', 'merkkiSelvakielinen'])['merkkiSelvakielinen'].size()
```

```python
data[(data.ajoneuvoluokka == 'M1')]['iskutilavuus'].mean()
```

```python
data['sylintereidenLkm'].value_counts()
```

```python
tyyppien_määrät = data.groupby(['ajoneuvoluokka']).size()
print(tyyppien_määrät)
tyyppien_määrät_descending = tyyppien_määrät.sort_values(ascending=False)
```

```python
#15 5) a) most common types... BAR-VERSION

plt.figure(num=None, figsize=(10, 5), dpi=100, facecolor='w', edgecolor='k')
plt.title('15 most common vehicle types in Finland')
plt.xlabel('Types')
plt.ylabel('Count')
plt.ylim([0,3e6])
plt.xlim([0, 14])

plt.bar(tyyppien_määrät_descending.index, tyyppien_määrät_descending.values)

#plt.show
```

```python
merkkien_määrät = data.groupby(['merkkiSelvakielinen']).size()
merkkien_määrät_descending = merkkien_määrät.sort_values(ascending=False)
print(merkkien_määrät_descending)
```

```python
#15 5) b) most common manufacturers... BAR-VERSION 

plt.figure(num=None, figsize=(20, 5), dpi=100, facecolor='w', edgecolor='k')
plt.title('15 most common vehicle types in Finland')
plt.xlabel('Types')
plt.ylabel('Count')
plt.ylim([0,5e5])
plt.xlim([0, 14])

plt.bar(merkkien_määrät_descending.index, merkkien_määrät_descending.values)

```
